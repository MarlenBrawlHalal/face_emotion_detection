```python
import tensorflow as tf
import os
import matplotlib.pyplot as plt
import numpy as np
import cv2

from tensorflow import keras
from keras import layers
from keras.preprocessing.image import image_utils #load_img, img_to_array
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
classes = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']

folder_path = '/content/drive/MyDrive/new_data/'

for cls in classes:
    path = os.path.join(folder_path, 'train', cls)
    lst = os.listdir(path)
    number_files = len(lst)
    print(cls, ': ', number_files)
```

```python
picture_size = 48
expression = 'disgust'

plt.figure(figsize=(12, 12))
for i in range(1, 10):
  plt.subplot(3, 3, i)
  img = image_utils.load_img(folder_path+'train/'+expression+'/'+
                             os.listdir(folder_path+'train/'+expression)[i],
                             target_size=(picture_size, picture_size))
  plt.imshow(img)
plt.show()
```

```python
img_size = 224
training_data = []

def create_training_data():
  for category in classes:
    path = os.path.join(folder_path, 'train', category)
    class_num = classes.index(category)
    for img in os.listdir(path):
      try:
        img_array = cv2.imread(os.path.join(path, img))
        new_array = cv2.resize(img_array, (img_size, img_size))
        training_data.append([new_array, class_num])
      except:
        pass
create_training_data()
```

```python
import random 

random.shuffle(training_data)
print(len(training_data))
```

```python
temp_array = np.array(training_data)
temp_array.shape
```

```python
X = []
y = []

for features, label in training_data:
  X.append(features)
  y.append(label)
X = np.array(X).reshape(-1, img_size, img_size, 3)
print(X.shape)
Y = np.array(y)
print(Y.shape)
X = X / 255.0
```

```python
model = tf.keras.applications.MobileNetV2()
model.summary()
```

```python
base_input = model.layers[0].input
print(base_input)

base_output = model.layers[-2].output

final_output = layers.Dense(128)(base_output)
final_output = layers.Activation('relu')(final_output)
final_output = layers.Dense(64)(final_output)
final_output = layers.Activation('relu')(final_output)
final_output = layers.Dense(7, activation='softmax')(final_output)
```

```python
new_model = keras.Model(inputs=base_input, outputs=final_output)

new_model.summary()

new_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam',
                     metrics=['accuracy'])
```

```python
new_model.fit(X, Y, epochs=30)
```

```python
new_model.save('facial_expression_model.h5')
```

```python
test_img = cv2.imread('happy.jpg')
img_bcp = test_img.copy()

plt.imshow(cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB))
```

```python
face_clf = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
gray_img = cv2.cvtColor(test_img, cv2.COLOR_BGR2GRAY)

faces_coordinates = face_clf.detectMultiScale(gray_img)

for (x, y, w, h) in faces_coordinates:
  cv2.rectangle(test_img, (x, y), (x + w, y + h), (0, 255, 0), 2)
  cropped_face = img_bcp[y:y+h, x:x+w]
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB))
```

```python
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(cropped_face, cv2.COLOR_BGR2RGB))
```

```python
class_dictionary = {0: 'angry', 1: 'disgust', 2: 'fear', 3: 'happy', 4: 'neutral', 5: 'sad', 6: 'surprise'}

final_img = cv2.resize(cropped_face, (224, 224))
final_img = np.expand_dims(final_img, axis=0)
final_img = final_img / 255.0

new_model = tf.keras.models.load_model('facial_expression_model.h5')
prediction = new_model.predict(final_img)
class_dictionary[np.argmax(prediction)]
```
